```python
dbutils.widgets.text("catalog_name", "dennis_schultz", "Catalog Name")
dbutils.widgets.text("schema_name", "sharepoint_testing", "Schema Name")
dbutils.widgets.text("volume_name", "test_documents", "Volume Name")

# Unity Catalog
catalog_name = dbutils.widgets.get("catalog_name")
schema_name = dbutils.widgets.get("schema_name")
volume_name = dbutils.widgets.get("volume_name")
```

```python
import shutil
import os

source_dir = "../SampleDocs"
target_dir = f"/Volumes/{catalog_name}/{schema_name}/{volume_name}"

os.makedirs(target_dir, exist_ok=True)

for filename in os.listdir(source_dir):
    src_file = os.path.join(source_dir, filename)
    dst_file = os.path.join(target_dir, filename)
    if os.path.isfile(src_file):
        shutil.copy2(src_file, dst_file)
```
